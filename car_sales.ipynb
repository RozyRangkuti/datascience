```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = df.drop(df[df['Engine_size'].isna() == True].index)
df.isna().sum()
```

```python
df[df['Price_in_thousands'].isna() == True]
```

```python
df['Price_in_thousands'] = df['Price_in_thousands'].fillna(df['Price_in_thousands'].mean())
df.isna().sum()
```

```python
df['Curb_weight'] = df['Curb_weight'].fillna(df['Curb_weight'].mean())
df.isna().sum()
```

```python
df['Fuel_efficiency'] = df['Fuel_efficiency'].fillna(df['Fuel_efficiency'].mean())
df['Power_perf_factor'] = df['Power_perf_factor'].fillna(df['Power_perf_factor'].mean())
df.isna().sum()
```

```python
df[df['__year_resale_value'].isna() == True]
```

```python
df[['Price_in_thousands', '__year_resale_value']]
```

```python
reduction_average = pd.Series(df['Price_in_thousands'] - df['__year_resale_value']).mean()
reduction_average
```

```python
df['__year_resale_value'] = df['__year_resale_value'].fillna(df['Price_in_thousands'].sub(reduction_average))
df.isna().sum()
```

```python
df['__year_resale_value']
```

```python
# Melihat model dengan penjualan mobil terbaik
df_model = df[['Model', 'Manufacturer', 'Sales_in_thousands']].groupby(['Model', 'Manufacturer']).sum().reset_index()
df_model = df_model.sort_values(['Sales_in_thousands'], ascending = False)
df_model.head()
```

```python
# Visualisasi model dengan penjualan mobil terbaik
fig = px.bar(df_model.head(10),
             x = 'Model',
             y = 'Sales_in_thousands',
             color = 'Model',
             text_auto = True,
             title = '10 Model dengan Jumlah Penjualan Terbaik')

fig.update_traces(textfont_size=10, textposition='inside')
fig.update_layout(width=1000, height=800)
fig.show()
```

```python
col_selected = ['Model', 'Vehicle_type', '__year_resale_value', 'Price_in_thousands', 'Engine_size', 'Horsepower', 'Wheelbase', 'Width', 'Length', 'Curb_weight', 'Fuel_capacity', 'Fuel_efficiency', 'Power_perf_factor']
```

```python
df_selected = df[col_selected].groupby(['Model', 'Vehicle_type']).mean().reset_index()
df_selected
```

```python
df_to_analyze = df_selected.merge(df_model, on='Model')
df_to_analyze.sort_values(['Sales_in_thousands'], ascending = False, inplace = True)
```

```python
df_to_analyze.head(10)
```

```python
fig = px.bar(df_to_analyze.head(10),
             x = 'Price_in_thousands',
             y = 'Model',
             color = 'Model',
             orientation = 'h',
             text_auto = True,
             title = 'Harga dari 10 model terbaik')

fig.update_traces(textfont_size = 10, textposition = "outside")
fig.show()
```

```python
# Angka horsepower yang dimiliki oleh model penjualan terbaik
fig = px.bar(df_to_analyze.head(10),
             x = 'Horsepower',
             y = 'Model',
             color = 'Model',
             orientation = 'h',
             text_auto = True,
             title = 'Angka Horsepower dari 10 model dengan penjualan terbaik')

fig.update_traces(textfont_size = 10, textposition = "outside")
fig.show()
```

```python
# Angka wheelbase yang dimiliki oleh model penjualan terbaik
fig = px.bar(df_to_analyze.head(10),
             x = 'Wheelbase',
             y = 'Model',
             color = 'Model',
             orientation = 'h',
             text_auto = True,
             title = 'Angka Wheelbase dari 10 model dengan penjualan terbaik')

fig.update_traces(textfont_size = 10, textposition = "outside")
fig.show()
```

```python
df_prediction = df_to_analyze.drop(['Model', 'Vehicle_type', '__year_resale_value', 'Price_in_thousands', 'Manufacturer', 'Sales_in_thousands'], axis = 1)
df_prediction.mean().round()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
pred_data = df_to_analyze.drop(['Model', 'Manufacturer', '__year_resale_value', 'Sales_in_thousands', 'Vehicle_type'], axis = 1)
pred_data
```

```python
x = pred_data.drop('Price_in_thousands', axis = 1)
y = pred_data['Price_in_thousands']
```

```python
x.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
model2 = Lasso()
model3 = Ridge()
model4 = LogisticRegression(multi_class='multinomial')
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
rmse = mean_squared_error(y_test, y_pred, squared = False)
r2score = r2_score(y_test, y_pred)
print('RMSE = ', rmse)
print('R2 Score = ', r2score)
```

```python
results_df = pd.DataFrame({'Actual Sales': y_test, 'Predicted Sales': y_pred})
results_df
```

```python
fig = px.scatter(results_df, x = 'Predicted Sales', y = 'Actual Sales', trendline='ols')
fig.update_layout(
    title = 'Actual vs Predicted Sales',
    xaxis_title = 'Predicted Sales',
    yaxis_title = 'Actual Sales'
)
fig.show()
```
