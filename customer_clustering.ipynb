```python
pred_data = df.drop(['id'], axis = 1)
pred_data
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
col_names = pred_data.columns
X = scaler.fit_transform(pred_data)
X = pd.DataFrame(data=X, columns=col_names)
X.describe()
```

```python
k_values = range(2, 12)
wcss = []
```

```python
for k in k_values:
  kmeans = KMeans(n_clusters=k, random_state=0)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
```

```python
fig = px.line(x = k_values, y = wcss, title = 'Elbow Plot')
fig.update_layout(xaxis_title='Number of Clusters (k)', yaxis_title='Inertia')
fig.show()
```

# Mereduksi Dimensional Data

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3, random_state = 42)
X_pca = pca.fit_transform(X)
```

```python
X_pca_df = pd.DataFrame(data=X_pca, columns=['X1', 'X2', 'X3'])
```

# Membuat Model Klasterisasi

```python
kmeans = KMeans(n_clusters=6, random_state=0)
kmeans.fit(X)
```

```python
labels = kmeans.labels_
X_pca_df['Labels'] = labels
```

```python
X_pca_df.head()
```

```python
# ubah tipe data menjadi string agar mempermudah dalam visualisasi
X_pca_df['Labels'] = X_pca_df['Labels'].astype(str)
```

```python
fig = px.scatter_3d(X_pca_df, x='X1', y='X2', z='X3', color=X_pca_df['Labels'])
fig.show()
```

```python
df['cluster'] = kmeans.labels_
df
```

# Analisa Karakteristik Hasil Klasterisasi

```python
# Jumlah data tiap klaster
df_cluster_n = df[['cluster', 'id']].groupby('cluster').count().sort_values('cluster').reset_index()
df_cluster_n
```

```python
# Visualisasi jumlah data tiap klaster
fig = px.bar(df_cluster_n, x='cluster', y='id', color='cluster', title='Perbandingan Jumlah Data Setiap Cluster', text_auto=True)
fig.update_traces(textfont_size=10, textposition='outside')
fig.show()
```
